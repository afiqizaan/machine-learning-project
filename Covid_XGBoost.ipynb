Handling Missing Values, Outliers and Type Casting

```python
#Replacing 97 and 99 with null
missing_values = (97,99)
df.replace(missing_values,np.nan, inplace=True)
df.head(100)
```

```python
# List of columns to exclude
exclude_columns = ['CLASIFFICATION_FINAL', 'USMER', 'MEDICAL_UNIT', 'DATE_DIED', 'AGE']

# Columns to be transformed
columns_to_transform = [col for col in df.columns if col not in exclude_columns]

# Transformation function
def transform_values(value):
    if value == 1:
        return 1  # yes
    elif value == 2:
        return 0  # no
    else:
        return value  # leave unchanged if not 1 or 2

# Apply the transformation
df[columns_to_transform] = df[columns_to_transform].applymap(transform_values)
# Display the updated DataFrame
df.head(100)   
```

```python
#Replacing the date 9999-99-99 with null
missing_values_date = "9999-99-99"
df.replace(missing_values_date, np.nan, inplace = True)
df.head(6)
```

```python
#Checking for missing values
df.isnull().sum().sort_values()
```

```python
# Missing values of DATE_DIED represents alive patients
# Missing values of AGE can be removed
# Missing values of PNEUMONIA, PREGNANT, INTUBED, and ICU should be kept because too much null value for each
df.dropna(subset=["AGE"], inplace=True)
```

```python
#Checking for outliers in "AGE" column
sns.boxplot(x=df["AGE"])
```

```python
#Removing the outliers in "AGE" features
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound =  Q3 + 1.5 * IQR
df = df[(df['AGE'] >= lower_bound) & (df['AGE'] <= upper_bound)]
sns.boxplot(x=df["AGE"])
```

```python
#Changing DATE_DIED string format to date format
df["DATE_DIED"] = pd.to_datetime(df["DATE_DIED"], format="%d/%m/%y", errors="ignore")
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Adjust class labels to start from 0 because XGBoost class start with 0
df['CLASIFFICATION_FINAL'] = df['CLASIFFICATION_FINAL'] - 1
```

```python
# Convert 'DATE_DIED' to a binary indicator: 0 if NaN, 1 otherwise
df['DATE_DIED'] = df['DATE_DIED'].notna().astype(int)
```

```python
#count number of people died 
death_counts = df['DATE_DIED'].value_counts()

sns.barplot(x=death_counts.index, y=death_counts.values, palette='viridis')
plt.title('Number of People Died vs Not Died')
plt.xlabel('Status (0: Not Died, 1: Died)')
plt.ylabel('Number of People')
plt.xticks([0, 1], ['Not Died', 'Died'])  
plt.ylim(0, death_counts.max() + 10)
plt.show()
```

```python
X = df.drop('CLASIFFICATION_FINAL', axis=1)
y = df['CLASIFFICATION_FINAL']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)
```

```python
xgb_y_pred = xgb_model.predict(X_test)

accuracy = accuracy_score(y_test, xgb_y_pred )
conf_matrix = confusion_matrix(y_test, xgb_y_pred )
class_report = classification_report(y_test, xgb_y_pred )
```

```python
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
